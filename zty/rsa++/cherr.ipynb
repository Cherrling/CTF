```python
# -*- coding:utf-8 -*-
#Author: Lazzaro
from Crypto.Util.number import getPrime,isPrime
from math import gamma
import random
```

```python
def nextPrime(n):
    n += 2 if n & 1 else 1
    while not isPrime(n):
        n += 2
    return n

def getNewPrime():
    A = getPrime(512)
    B = nextPrime(A - random.randint(10000,100000))
    print("A = ",A)
    print("B = ",B)
    print(gamma(B))
    return nextPrime(gamma(B+2)%A)
p=getNewPrime()
print("p = ",p)
```
