```python
#访问指定的地址，获取返回的内容
import requests
import re
import time
import json
url = 'http://119.45.42.24:20025/'

#获取返回的内容
def get_content():
    response = requests.get(url)
    content = response.text
    # 获取header中的set-cookie
    cookie = response.headers['set-cookie']
    print(content)
    return content , cookie

# 从中解析出算式
def get_expression(content):
    # 分离等号前的算式
    expression = content.split('=')[0]
    print(expression)
    return expression


# 获取返回内容中setcookie的值
def get_cookie(content):
    cookie = re.findall(r'set-cookie: (.*?) ', content)
    print(f'cookie: {cookie}')
    return cookie[0]

# 设置cookie
def set_cookie(cookie):
    headers = {
        'Cookie': cookie
    }
    return headers

# 使用POST向另一个地址发送答案
def send_answer(calc,answer, headers):
    url = 'http://119.45.42.24:20025/admin/calc'
    data = {
        'calc': calc,
        'answer': answer
    }
    json_data=json.dumps(data)
    print(json_data)
    response = requests.post(url, data=json_data, headers=headers)
    return response.text

# 主函数
def main():
    while True:
        content,cookie = get_content()
        expression = get_expression(content)
        print(f'expression: {expression}')
        answer = eval(str(expression))
        print(f'answer: {answer}')
        print(f'cookie: {cookie}')
        headers = set_cookie(cookie)
        result = send_answer(expression,answer, headers)
        print(result)
        
main()
```
