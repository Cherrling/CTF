```python
# 使用 Wiener's attack
def wiener_attack(e, n):
    def convergents_of_e_over_n(e, n):
        convergents = []
        a, b = e, n
        while b:
            convergents.append(a // b)
            a, b = b, a % b
        return convergents

    convergents = convergents_of_e_over_n(e, n)
    for k, d in enumerate(convergents):
        if k == 0:
            continue
        phi = (e * d - 1) // k
        b = n - phi + 1
        delta = b * b - 4 * n
        if delta < 0:
            continue
        if delta == 0:
            p = b // 2
            q = n // p
            if p * q == n:
                return p, q
        else:
            delta_sqrt = gmpy2.isqrt(delta)
            if delta_sqrt * delta_sqrt == delta:
                p = (b + delta_sqrt) // 2
                q = (b - delta_sqrt) // 2
                if p * q == n:
                    return p, q
    return None
```
