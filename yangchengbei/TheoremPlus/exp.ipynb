```python
def decode_e(e):
    if e > 1:
        mul = 1
        for i in range(1, e):
            mul *= i
        if e - mul % e - 1 == 0:
            mulmod = mul % e - e
        else:
            mulmod = mul % e
        return mulmod + decode_e(e - 1)
    else:
        return 0
e = abs(decode_e(703440151))
print(e)
```

```python
#埃拉托斯特尼筛法统计1-p之间的素数个数
def prime_pi(p):
    prime = [1] * (p+1)
    prime[0] = prime[1] = 0
    for i in range(2, int(p**0.5)+1):
        if prime[i] == 1:
            prime[i*i:p+1:i] = [0] * len(prime[i*i:p+1:i])
    return sum(prime)

e = prime_pi(703440151) - 2
print(f'e = {e}')
# e = 36421873
```
